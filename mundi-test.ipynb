```python
import mundi
import pandas as pd

names = mundi.regions_dataframe(country_id='BR', type='state')
capitals = [*map(lambda x: f'BR-{x}', [1100205,1200401,1302603,1400100,1501402,1600303,1721000,2111300,2211001,2304400,2408102,2507507,2611606,2704302,2800308,2927408,3106200,3205309,3304557,3550308,4106902,4205407,4314902,5002704,5103403,5208707,5300108])]
names = pd.DataFrame([], index=[*names.index, *capitals]).mundi[['name']]

names = names.astype('string')
names['name'] = names['name'] + (names.mundi['type'] == 'city').apply(lambda x: ' (capital)' if x else '')
names.columns = pd.MultiIndex.from_tuples([('name', '')])

ages = pd.concat({'age_distribution': names.mundi['age_distribution']}, axis=1)
```

```python
data = pd.concat([names, ages], axis=1)
data = pd.concat([
    data,
    pd.DataFrame([
        ['Brasil', *data['age_distribution'].iloc[:len(data) // 2].sum()]
    ], columns=data.columns, index=['BR'])
])
data.to_pickle('data.pkl.gz')
data
```
